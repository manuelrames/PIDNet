# 1. Register Railways SemSegm dataset to AzureML
## 1.1. View datastores

Run the following code to determine the datastores in your workspace:

```python
from azureml.core import Workspace

ws = Workspace.from_config()
print(ws.name, "loaded")

# Get the default datastore
default_ds = ws.get_default_datastore()

# Enumerate all datastores, indicating which is the default
for ds_name in ws.datastores:
    print(ds_name, "- Default =", ds_name == default_ds.name)
```

## 1.2. Upload dataset folder from local and register dataset

```python
from azureml.core import Dataset

# upload the data to datastore and create a FileDataset from it 
folder_data = Dataset.File.upload_directory(src_dir="../railways_semsegm_5classes_dataset", target=(default_ds, "azureml/railways_semsegm_5classes_dataset"))
dataset = folder_data.register(workspace=ws, name="railways_semsegm_5classes_dataset")
```
